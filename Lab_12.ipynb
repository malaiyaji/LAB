# MATRIX CHAIN MULTIPLICATION Dynamic Programming with table filling.
Problem Description:
Find minimum number of scalar multiplications needed to multiply a chain of matrices..

#Theory
Cost of multiplying two matrices:
If A is p×q and B is q×r, cost = p × q × r

Dynamic Programming idea:
Break problem into smaller subproblems.
Store results in a table to avoid recomputation.

m[i][j] = minimum cost to multiply Ai to Aj

## Algorithm
1. Take array p[] of size n
2. Create table m[n][n]
3. Set m[i][i] = 0
4. Fill table diagonally
5. For each chain length L:
    Compute m[i][j]
    Try all possible k between i and j
6. Choose minimum cost
7. Return m[1][n-1]

```python
import sys

def matrix_chain_multiplication(p):
    n = len(p)

    # Create DP table for costs
    m = [[0 for _ in range(n)] for _ in range(n)]
    # Create table for split points to reconstruct optimal parenthesization
    s = [[0 for _ in range(n)] for _ in range(n)]

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                # Cost to multiply matrices Ai to Ak and Ak+1 to Aj,
                # plus cost to multiply the two resulting matrices.
                cost = (m[i][k] +
                        m[k+1][j] +
                        p[i-1] * p[k] * p[j])

                if cost < m[i][j]:
                    m[i][j] = cost
                    s[i][j] = k # Store the split point

    return m, s # Return both tables
```

```python
# Input dimensions for the initial example
p = [10, 30, 5, 60]


table, s_table = matrix_chain_multiplication(p) # Now returns both cost and split tables

print("Minimum multiplication cost:", table[1][len(p)-1])

print("\nDP Cost Table:")
for row in table[1:]:
    print(row[1:])

print("\nDP Split Table (s):")
for row in s_table[1:]:
    print(row[1:])
```

```python
def print_optimal_parenthesization(s, i, j):
    if i == j:
        return f"A{i}"
    else:
        k = s[i][j]
        return f"({print_optimal_parenthesization(s, i, k)}{print_optimal_parenthesization(s, k + 1, j)})"
```

```python
def calculate_naive_cost(p):
    n = len(p)
    if n < 3: # Need at least two matrices to multiply
        return 0

    current_p = p[0] # Dimension for the first matrix's rows
    current_q = p[1] # Dimension for the first matrix's columns
    total_cost = 0

    # Multiply A1 to A2, then (A1A2) to A3, and so on.
    # This is equivalent to left-to-right parenthesization.
    for i in range(1, n - 1):
        # Current matrices being multiplied: (result of previous) x A_i
        # Dimensions for current product: (current_p x current_q) x (current_q x p[i+1])
        # Cost for this step is current_p * current_q * p[i+1]
        total_cost += current_p * current_q * p[i+1]
        current_q = p[i+1] # The new 'q' becomes the columns of the next matrix
    return total_cost
```

## Exercise Solution

```python
# 1. Given p = [5, 10, 3, 12, 5, 50, 6]
exercise_p = [5, 10, 3, 12, 5, 50, 6]

# 2. Compute minimum cost
exercise_m_table, exercise_s_table = matrix_chain_multiplication(exercise_p)

min_cost = exercise_m_table[1][len(exercise_p) - 1]
print(f"Minimum multiplication cost for p = {exercise_p}: {min_cost}")

# 3. Print optimal parenthesization
# Note: The print_optimal_parenthesization function assumes 1-indexed matrices for clarity (A1, A2, etc.)
# but the s_table uses 0-indexed split points. Adjusting for 1-indexing for output.
optimal_parenthesization = print_optimal_parenthesization(exercise_s_table, 1, len(exercise_p) - 1)
print(f"Optimal parenthesization: {optimal_parenthesization}")

# 4. Compare number of multiplications saved
naive_cost = calculate_naive_cost(exercise_p)
saved_multiplications = naive_cost - min_cost

print(f"Naive (left-to-right) multiplication cost: {naive_cost}")
print(f"Multiplications saved (naive - optimal): {saved_multiplications}")

print("\nDP Cost Table (exercise):")
for row in exercise_m_table[1:]:
    print(row[1:])

print("\nDP Split Table (exercise):")
for row in exercise_s_table[1:]:
    print(row[1:])
```
