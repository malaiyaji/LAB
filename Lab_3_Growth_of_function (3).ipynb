# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Prob_2: Another example of O(1) Constant Time Complexity

def is_even(number):
    return number % 2 == 0

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    is_even(n)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Is Even Check")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Prob_3: Another example of O(n) Linear Time Complexity

def count_greater_than_zero(numbers):
    count = 0
    for num in numbers:
        if num > 0:
            count += 1
    return count

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = [i - n//2 for i in range(n)]
    count_greater_than_zero(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity for Counting Positive Elements")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.
import matplotlib.pyplot as plt
import math

def binary_search(arr, target):
    operations = 0
    left, right = 0, len(arr) - 1
    while left <= right:
        operations += 1
        mid = (left + right) // 2
        if arr[mid] == target:
            return operations
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return operations

n_values = []
time_values = []

for i in range(1, 15):
    n = 2**i
    arr = list(range(n))

    if n > 0:
        n_values.append(n)
        time_values.append(math.log2(n) if n > 1 else 1)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-')
plt.xlabel("Input size n (2^i)")
plt.ylabel("Time (log n)")
plt.title("O(log n) Logarithmic Time Complexity (Binary Search)")
plt.xscale('log')
plt.grid(True)
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.
import matplotlib.pyplot as plt
import math
import time

def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    return merge(left_half, right_half)

def merge(left, right):
    merged = []
    left_idx, right_idx = 0, 0

    while left_idx < len(left) and right_idx < len(right):
        if left[left_idx] < right[right_idx]:
            merged.append(left[left_idx])
            left_idx += 1
        else:
            merged.append(right[right_idx])
            right_idx += 1

    merged.extend(left[left_idx:])
    merged.extend(right[right_idx:])
    return merged

n_values = []
time_values = []

for i in range(1, 15):
    n = 2**i
    n_values.append(n)
    if n > 1:
        time_values.append(n * math.log2(n))
    else:
        time_values.append(0)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-')
plt.xlabel("Input size n (2^i)")
plt.ylabel("Time (n log n)")
plt.title("O(n log n) Log-Linear Time Complexity (Merge Sort)")
plt.grid(True)
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in polynomial time (e.g., O(n^2)). plot the graph as well, similar to above example

import matplotlib.pyplot as plt
def polynomial_time_example(n):
    count = 0
    for i in range(n):
        for j in range(n):
            count += 1
    return count

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    time_values.append(n**2)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-')
plt.xlabel("Input size n")
plt.ylabel("Time (n^2)")
plt.title("O(n^2) Polynomial Time Complexity")
plt.grid(True)
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in exponential time (e.g., O(c^n)). plot the graph as well, similar to above example
import matplotlib.pyplot as plt

def fibonacci_recursive(n):
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

n_values = list(range(1, 20))
time_values = []

for n in n_values:
    time_values.append(2**n)

plt.figure(figsize=(10, 6))
plt.plot(n_values, time_values, marker='o', linestyle='-')
plt.xlabel("Input size n")
plt.ylabel("Time (2^n)")
plt.title("O(2^n) Exponential Time Complexity (e.g., Recursive Fibonacci)")
plt.grid(True)
plt.yscale('log')
plt.show()
```

```python
# Write a function to perform quick sort and show that it has Logarithmic Time O(log n) Complexity
import matplotlib.pyplot as plt
import math
import random
import time

def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

n_values = list(range(1, 1001, 50))
time_values_quicksort = []
time_values_nlogn_theoretical = []

for n in n_values:
    data = [random.randint(0, 10000) for _ in range(n)]

    start_time = time.time()
    quicksort(data)
    end_time = time.time()
    time_values_quicksort.append((end_time - start_time) * 1000)

    if n > 0:
        time_values_nlogn_theoretical.append(n * math.log2(n))
    else:
        time_values_nlogn_theoretical.append(0)

max_actual_time = max(time_values_quicksort) if time_values_quicksort else 1
max_theoretical_val = max(time_values_nlogn_theoretical) if time_values_nlogn_theoretical else 1

scaling_factor = max_actual_time / max_theoretical_val
normalized_theoretical_nlogn = [val * scaling_factor for val in time_values_nlogn_theoretical]

plt.figure(figsize=(12, 8))

plt.plot(n_values, time_values_quicksort, label='Quicksort (Actual Time)', marker='o', linestyle='-')
plt.plot(n_values, normalized_theoretical_nlogn, label='O(n log n) Theoretical', linestyle='--', color='red')

plt.xlabel('Input size n')
plt.ylabel('Time (milliseconds)')
plt.title('Quicksort Performance: Demonstrating O(n log n) Complexity')
plt.legend()
plt.grid(True)
plt.show()
```
